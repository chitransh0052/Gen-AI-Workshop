```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
#load the tokenizer and model
# medium is a good balance between performance and speed. you can also try 'small orlarge,
tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-medium")
model=AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-medium")

```

```python
#chat history will be stored in this variable
chat_history_ids = None
```

```python
print("Hey !Chatbot is ready! Type 'quit' to stop.\n")
# Loop for chatting
for step in range(100):  # Limit to 100 exchanges to avoid infinite loop
    # Get user input
    user_input = input("You: ")

    if user_input.lower() in ['quit', 'exit', 'bye']:
        print(" Chatbot: Goodbye!")
        break

    # Encode the user input and add end-of-string token
    new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')

    # Append the new user input to the chat history (for context)
    if chat_history_ids is not None:
        # Append new input to the existing history
        input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1)
    else:
        input_ids = new_input_ids  # First turn

    # Generate a response from the model
    chat_history_ids = model.generate(
        input_ids,
        max_length=1000,             # Max tokens in the conversation
        pad_token_id=tokenizer.eos_token_id,
        do_sample=True,              # Sampling for randomness
        top_k=50,                    # Top-k sampling
        top_p=0.95,                  # Nucleus sampling
        temperature=0.7              # Controls randomness
    )

    # Extract only the new response (exclude previous turns)
    response = tokenizer.decode(chat_history_ids[:, input_ids.shape[-1]:][0], skip_special_tokens=True)

    print(f"Chatbot: {response}")
```
